```python
import pandas as pd

dia = pd.read_csv("C:/Users/silan/Python/Data/diamonds.csv")
dia.head()
```

```python
pd.crosstab(dia['cut'], dia['clarity'])
```

```python
pd.crosstab(dia['cut'], dia['clarity'], normalize = True).round(2)
```

```python
# row에 대한 normalize 연산 시 normalize = 0

pd.crosstab(dia['cut'],dia['clarity'], normalize = 0)
```

```python
# column에 대한 normalize 연산 시 normalize = 1

pd.crosstab(dia['cut'],dia['clarity'], normalize = 1)
```

```python
pd.crosstab(dia['cut'],dia['clarity'], values = dia['price'],
           aggfunc = pd.Series.mean) # 집계 방식 평균
```

```python
cross_t = pd.crosstab(dia['cut'], dia['color'])
cross_t
```

```python
cross_t.reset_index()
```

```python
cross_melt = cross_t.reset_index().melt(id_vars = 'cut')
cross_melt.head()
```

```python
cross_melt.sort_values("value", ascending = False).head(2)
```

```python
# groupby를 활용한 동일한 평균 집계 표현

dia.groupby(['cut', 'clarity'])['price'].mean().reset_index()
```

```python
dia_agg = dia.groupby(['cut', 'color'])[['price', 'carat']].mean().reset_index()
dia_agg.head()
```

```python
dia_agg['ratio'] = dia_agg['price']/ dia_agg['carat'] # 1캐럿 당 가격
dia_agg.sort_values(["ratio"], ascending = False)
```

```python
dia_agg.sort_values("price", ascending = False).head()
```

```python
dia_agg.sort_values(['cut', 'clarity'], ascending = [True, False]).head() # 정렬 기준이 2개 이상일 때
```

### Pandas_melt
- wide form의 데이터 프레임을 long form으로 자료구조를 변환하는 메서드
- id_vars 인자에 기준이 되는 변수를 지정하여 처리
- 통계 또는 머신러닝 실시 전에 데이터 구조를 맞추기 위해 주로 사용

```python
# 실습데이터: electricity usage

elec = pd.read_csv("C:/Users/silan/Python/Data/elec_load_2017_7d.csv")
elec
```

```python
# id_vars: 식별자 변수

elec_melt = elec.melt(id_vars = ['YEAR','MONTH','DAY'])
elec_melt.head(3)
```

```python
print(len(elec))
print(len(elec_melt))
```

### Pandas_pivot()
- long form의 데이터 프레임을 wide form으로 자료 구조를 변환하는 메서드 (`melt`와 반대)
- `pd.crosstab()` 보다 더 자주 활용되며, index/columns/values인자에 각각 대상 변수를 지정하여 출력데이터 구조 결정
- 데이터를 요약하거나 군집분석 실시 전 데이터 전처리에 주로 활용

```python
# 상세 집계 방식까지 추가하려면 pivot_table()
elec_pivot = elec_melt.pivot(index = ['YEAR','MONTH','DAY'], 
                             columns = 'variable', 
                             values = 'value').reset_index()

elec_pivot
```
