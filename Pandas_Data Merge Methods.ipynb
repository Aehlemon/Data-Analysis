### 데이터 전처리_데이터 병합
1. 병합의 종류
- 데이터를 이어 붙이는 바인딩 (binding)은 행/열 방향으로 연산가능
- 특정변수를 기준으로 2개의 데이터를 병합하는 `join`을 많이 사용
- 조인 연산 중 `Inner Join`과 `Left Join`을 많이 사용
 - `Inner Join` : 2개 데이터 세트의 기준이 되는 변수인 'Key' 변수가 공통인 행을 반환
 - `Left Join`: 2개 데이터 세트 중 첫번째 세트의 모든 행을 유지하고 두번째 세트는 Key 변수 기준으로 첫번째 세트와 공통인 행을 반환

<br>

#### 비가 온 날 습도가 100일 때 온도가 30도 넘는 시각의 count의 평균??
- humidity의 최댓값이 100인 날을 비가 온 날로 간주한다.
- 반올림하여 둘째 자리까지 표기

```python
import pandas as pd
bike = pd.read_csv("C:/Users/silan/Python/Data/bike.csv")

bike["datetime"] = pd.to_datetime(bike["datetime"])
bike["date"] = bike["datetime"].dt.date # date(yyyy-mm-dd)을 불러오는 메소드
bike.head(5)
```

```python
bike_h100 = bike.groupby(['date'])['humidity'].max().reset_index()
bike_h100.head(5)
```

```python
bike_h100 = bike_h100.loc[bike_h100['humidity'] == 100]
bike_h100.head(3)
```

```python
len(bike_h100)
```

```python
bike_join = pd.merge(left = bike, right = bike_h100,
                    left_on = "date", right_on = "date",
                    how = "inner")
bike_join.head(4)
```

```python
bike_join_temp_30 = bike_join.loc[bike_join['temp'] > 30]
bike_join_temp_30
```

```python
bike_join_temp_30['count'].mean()
```
