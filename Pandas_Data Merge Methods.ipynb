### 데이터 전처리_데이터 병합
1. 병합의 종류
- 데이터를 이어 붙이는 바인딩 (binding)은 행/열 방향으로 연산가능
- 특정변수를 기준으로 2개의 데이터를 병합하는 `join`을 많이 사용
- 조인 연산 중 `Inner Join`과 `Left Join`을 많이 사용
 - `Inner Join` : 2개 데이터 세트의 기준이 되는 변수인 'Key' 변수가 공통인 행을 반환
 - `Left Join`: 2개 데이터 세트 중 첫번째 세트의 모든 행을 유지하고 두번째 세트는 Key 변수 기준으로 첫번째 세트와 공통인 행을 반환

<br>
2. 주요 함수 및 메소드: 
- `reset_index()`; 인덱스 초기화, 0부터 row 개수 만큼 숫자가 재할당 (자동~)
- `set_index()`;  특정 변수를 인덱스로 지정할 경우 사용 (데이터 병합 또는 시계열 분해에서 연산을 위해 활용)

```python
# Concat 합치기

bike_1 = bike.iloc[:3, :4]
bike_2 = bike.iloc[5:8, :4]
```

```python
bike_1.shape
```

```python
bike_2.shape
```

```python
print(bike_1)
print(bike_2)
```

```python
bike_merge = pd.concat([bike_1, bike_2]) # 행으로 이어 붙이기
bike_merge
```

```python
# 열로 붙여보면,
pd.concat([bike_1, bike_2], axis = 1) # 데이터 값이 안 맞아서 nan 생김
```

```python
pd.concat([bike_1.reset_index(drop = True), 
           bike_2.reset_index(drop = True)], axis = 1) # 강제로 할 수 있음
```
